```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.linear_model import LogisticRegression
import numpy as np
import joblib
```

```python
IMG_SIZE = 256
BATCH_SIZE = 32
```

```python
train_datagen = ImageDataGenerator(rescale=1./255, validation_split=0.2)
```

```python
train_generator = train_datagen.flow_from_directory('Hurricane_Damage/train_another',
                                                    target_size=(IMG_SIZE, IMG_SIZE),
                                                    batch_size=BATCH_SIZE,
                                                    class_mode='binary',
                                                    subset='training')
```

```python
validation_generator = train_datagen.flow_from_directory('Hurricane_Damage/train_another',
                                                         target_size=(IMG_SIZE, IMG_SIZE),
                                                         batch_size=BATCH_SIZE,
                                                         class_mode='binary',
                                                         subset='validation')
```

```python
#Feature extraction
def extract_features(gen):
    X, y = [], []
    for i in range(len(gen)):
        imgs, labels = gen[i]
        X.append(imgs.reshape(imgs.shape[0], -1))
        y.append(labels)
        return np.vstack(X), np.hstack(y)

X_train, y_train = extract_features(train_generator)
test_X, test_y = extract_features(validation_generator)
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
joblib.dump(model, "hurricane_classifier.pkl")
```

```python
prediction = model.predict(test_X)
for value in prediction:
    if value > 0.5:
        print('Hurricane Damage Detected')
    else:
        print('Hurricane Damage Not Detected')
```
